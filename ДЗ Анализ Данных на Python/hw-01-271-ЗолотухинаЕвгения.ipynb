### Задача 1 (Парадокс инспектора)

Автобусы отходят от автостанции с 8:00 до 20:00. Первый автобус отходит ровно в 8:00. Затем интервалы между автобусами случайны, независимы и равновероятно равны либо 5-и, либо 10-и минутам. Будем считать, что за минуту на автостанцию приходит ровно один пассажир, и все пассажиры едут ближайшим автобусом.

Проведите $10^4$ экспериментов и с их помощью:

__а) [5 баллов]__ Постройте гистограмму количества автобусов, отошедших от автостанции за сутки. Похоже ли визуально распределение на нормальное?

__б) [10 баллов]__ Инспектор Тимон выбирает равновероятно один из всех автобусов отошедших от автостанции за сутки. Постройте гистограмму числа пассажиров на этом автобусе. Оцените математическое ожидание и дисперсию этого числа.

__в) [10 баллов]__ Инспектор Пумба приходит на автостанцию в случайный момент времени, равномерный от 8:00 до 20:00 и садится в первый пришедший автобус. Постройте гистограмму числа пассажиров на этом автобусе. Оцените математическое ожидание и дисперсию этого числа.

__г) [10 баллов]__ Как изменятся ответы на эти вопросы, если время между автобусами будет экспоненциально со средним в 10 минут?

__Пояснения:__ первый автобус можно не включать в выборку, Пумбу за пассажира можно не считать. 

__Пункт а:__

```python
import random
import numpy as np 
import matplotlib.pyplot as plt
from collections import Counter
from scipy import stats
mins_delta = 12 * 60
n_obs = 10**4
np.random.seed(0)

ans = np.random.choice([5, 10], p = [0.5, 0.5], size = (n_obs, 144))
sum_ans = np.cumsum(ans, axis = 1)
buses = np.eye(n_obs, 1)
for i in range(n_obs):
    for j in range(144):
        if (sum_ans[i][j] <= 720):
            buses[i][0] += 1
            
x = buses.tolist()
res = dict(sum(map(Counter, x), Counter()))
plt.bar(list(res.keys()), res.values(), color='b')
plt.ylabel('Количество экспериментов')
plt.xlabel('Количество автобусов');
```

__Вывод:__ Данное распределение визуально похоже на нормальное. Распределение вероятностей практически симметрично относительно среднего значения(96 автобусов в день), это показывает, что данные около среднего значения встречаются чаще, чем данные, далекие от среднего значения.

### Задача 2 (Парадокс Хуана Мануэля Родригеса Паррондо)

У Атоса, Портоса и Арамиса по 1000 франков. Атос постоянно ходит в казино А, где каждый раз выигрывает один франк с вероятностью $0.49$ и проигрывает один франк с вероятностью $0.51$. Портос ходит в казино Б, где ситуация интереснее :) 

- Если богатство посетителя кратно трём, то посетитель выигрывает франк с вероятностью $0.09$ и проигрывает один франк с вероятностью $0.91$
- Если богатство посетителя не кратно трём, то посетитель выигрывает франк с вероятностью $0.74$ и проигрывает один франк с вероятностью $0.26$

Арамис каждый раз выбирает казино А или казино Б равновероятно. Проведите $10^4$ симуляций эволюции благосостояния Атоса, Портоса и Арамиса.Каждая симуляция предусматривает $1000$ посещений казино. Все три игрока все деньги носят с собой и ни на что не тратят :)

__а) [10 баллов]__ Постройте на графике $10$ случайных траекторий изменений богатства Атоса, по горизонтали — номер посещени казино, по вертикали — богатство. На том же графике постройте усреднённую по всем экспериментам тракторию изменения богатства.

__б) [10 баллов]__ Постройте на графике $10$ случайных траекторий изменений богатства Портоса, по горизонтали —  номер посещени казино, по вертикали — богатство. На том же графике постройте усреднённую по всем экспериментам тракторию изменения богатства.

__в) [10 баллов]__ Постройте на графике $10$ случайных траекторий изменений богатства Арамиса, по горизонтали — номер посещени казино, по вертикали — богатство. На том же графике постройте усреднённую по всем экспериментам тракторию изменения богатства.
    

__Пункт а__

```python
n_obs = 10**4
np.random.seed(0)
x = np.random.choice([1, -1], p = [0.49, 0.51], size = (n_obs, 1000))
wealth_Atos = np.full((n_obs, 1000), 1000)
wealth_change_Atos = np.cumsum(x, axis = 1)
trajectory_Atos = wealth_Atos + wealth_change_Atos
random = trajectory_Atos[(np.random.choice(trajectory_Atos.shape[0], 10, replace=False)), :] 
indices = np.full((10, 1000), range(1, 1001))
indices_2 = np.array(range(1, 1001))
for m in range(10):
    plt.plot(indices[m][:], random[m][:], linewidth = 0.5)
    plt.ylabel('Богатство Атоса')
    plt.xlabel('Номер посещения казино');
plt.plot(indices_2, trajectory_Atos.mean(axis = 0), color = "k", label = "Усреднённое богатство", linewidth = 2) 
plt.legend(loc='upper left');
```

__Вывод:__ Играя в игру, в которой стратегия не зависит от суммы капитала игрока, траектория изменения его богастства будет сильно колебаться, и в большинстве случаев игрок будет терять деньги, то есть проигрывать (Это показывают цветные линии графика). Усреднённое богатство (чёрная линия) показывает, что, действительно, играя в данную игру много раз подряд, игрок в итоге останется в проигрыше (с увеличением числа походов в казино богатство Атоса уменьшается).

__Пункт б__

```python
np.random.seed(0)
wealth_Portos = np.full((n_obs, 1000), 1000)
for i in range(n_obs):
    for j in range(1000):
        if wealth_Portos[i][j] % 3 == 0:
            wealth_Portos[i][j:] += np.random.choice([1, -1], p = [0.09, 0.91])
        else: 
            wealth_Portos [i][j:] += np.random.choice([1, -1], p = [0.74, 0.26])
random_Portos = wealth_Portos[(np.random.choice(wealth_Portos.shape[0], 10, replace=False)), :]   
for m in range(10):
    plt.plot(indices[m][:], random_Portos[m][:], linewidth = 0.5)
    plt.ylabel('Богатство Портоса')
    plt.xlabel('Номер посещения казино');
plt.plot(indices_2, wealth_Portos.mean(axis = 0), color = "k", label = "Усреднённое богатство", linewidth = 2) 
plt.legend(loc='upper left');
```

__Вывод:__ Играя в игру, в которой стратегия зависит только от суммы капитала игрока, в траектории изменения его богастства наблюдаются не столь сильные колебания, как в предыдущем случае. Однако всё так же в большинстве случаев игрок будет терять деньги, то есть проигрывать (Это показывают цветные линии графика). Усреднённое богатство (чёрная линия) показывает, что, действительно, играя в данную игру много раз подряд, игрок в итоге останется в проигрыше (с увеличением числа походов в казино богатство Портоса уменьшается).

__Пункт в__

```python
np.random.seed(0)
wealth_Aramis = np.full((n_obs, 1000), 1000)
for i in range(n_obs):
    for j in range(1000):
        casino = np.random.choice([0, 1], p = [0.5, 0.5])
        if casino == 0:
            wealth_Aramis[i][j:] += np.random.choice([1, -1], p = [0.49, 0.51])
        else:
            if wealth_Aramis[i][j] % 3 == 0:
                wealth_Aramis[i][j:] += np.random.choice([1, -1], p = [0.09, 0.91])
            else: 
                wealth_Aramis [i][j:] += np.random.choice([1, -1], p = [0.74, 0.26])
random_Aramis = wealth_Aramis[(np.random.choice(wealth_Aramis.shape[0], 10, replace=False)), :]   
for m in range(10):
    plt.plot(indices[m][:], random_Aramis[m][:], linewidth = 0.5)
    plt.ylabel('Богатство Арамиса')
    plt.xlabel('Номер посещения казино');
plt.plot(indices_2, wealth_Aramis.mean(axis = 0), color = "k", label = "Усреднённое богатство", linewidth = 2) 
plt.legend(loc='upper left');
```

__Вывод:__ Играя в игру, в которой стратегия периодически меняется в зависимости от выбора казино, траектория изменения богатства игрока может иметь сильные колебания, однако в большинстве случаев по окончании игры игрок будет увеличивать своё богатство, то есть выигрывать (Это показывают цветные линии графика). Усреднённое богатство (чёрная линия) показывает, что сложная, комбинированная игра может оказаться выигрышной, то есть в ней игрок при длительной игре выигрывает (с увеличением числа походов в казино богатство Арамиса увеличилось).

### Задача 3 (Парадокс Берксона)

Предположим, что результаты ЕГЭ школьников по русскому и математике независимы и хорошо аппроксимируются 
нормальным распределением с ожиданием 60 и стандартным отклонением 10. 

УШЭ (Урюпинская Школа Экономики) ныне престижна и забирает себе всех школьников набравших более $n$ баллов в сумме по русскому и математике. 

Для каждого $n$ от $80$ до $160$ с шагом $5$ случайно создайте $10^4$ школьников и поделите их на прошедших и не прошедших в УШЭ.


__а) [5 баллов]__ Постройте график числа прошедших в УШЭ в зависимости от $n$. 

__б) [5 баллов]__ Постройте график выборочной корреляции между результатами по русскому и математике в зависимости от $n$ среди прошедших в УШЭ.

__в) [5 баллов]__ Постройте график выборочной корреляции между результатами по русскому и математике в зависимости от $n$ среди не прошедших в УШЭ.

__Пункт а__

```python
np.random.seed(0)
norm_rvs = np.random.normal(loc=60, scale=10, size = (2, n_obs))
total_score = norm_rvs.sum(axis = 0)
a = []
for n in range (80, 161, 5):
    a.append(np.count_nonzero (total_score >= n ))
plt.plot((np.arange(80, 161, 5)), np.array(a), color="b", marker='o')
plt.ylabel('Число прошедших в УШЭ')
plt.xlabel('n');
```

__Вывод:__ С увеличением проходного балла (n) число школьников, прошедших в УШЭ, уменьшается, то есть наблюдается обратная зависимость между двумя признаками.

__Пункт б__

```python
coef = []
for n in range (80, 161, 5):
    b = np.where (total_score >= n )
    c = [norm_rvs[0][b].tolist()]
    d = [norm_rvs[1][b].tolist()]
    for i, e in zip(c, d):
        coef.append(stats.pearsonr(i, e)[0])  
plt.plot((np.arange(80, 161, 5)), np.array(coef), color="b", marker='o')
plt.title('График выборочной корреляции между результатами по русскому и математике среди прошедших в УШЭ')
plt.ylabel('Коэффициент корреляции Пирсона')
plt.xlabel('n');      
```

__Вывод:__ Для каждого проходного балла (n) наблюдается отрицательная линейная корреляция между результатами по русскому и математике среди прошедших в УШЭ. Чем больше значение n, тем дальше коэффициент корреляции от 0, то есть разброс данных становится меньше, и взаимосвязь между результатами становится сильнее.

__Пункт в__

```python
coef_2 = []
for n in range (80, 161, 5):
    res = np.where (total_score < n )
    math = [norm_rvs[0][res].tolist()]
    rus = [norm_rvs[1][res].tolist()]
    for i, e in zip(math, rus):
        coef_2.append(stats.pearsonr(i, e)[0])  
plt.plot((np.arange(80, 161, 5)), np.array(coef_2), color="b", marker='o')
plt.title('График выборочной корреляции между результатами по русскому и математике среди не прошедших в УШЭ')
plt.ylabel('Коэффициент корреляции Пирсона')
plt.xlabel('n');  
```

__Вывод:__ Для каждого проходного балла (n) наблюдается отрицательная линейная корреляция между результатами по русскому и математике среди прошедших в УШЭ. Чем больше значение n, тем ближе коэффициент корреляции к 0, то есть разброс данных становится больше, и взаимосвязь между результатами становится слабее.
